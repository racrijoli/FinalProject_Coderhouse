```python
# Importando bibliotecas

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport

# Estilo dos gr√°ficos
mpl.style.use("ggplot")
```

# Preparando os dados.

```python
# Importando a base

data = pd.read_csv("/content/ESS10.csv")
data.head()
```

```python
# Filtrando as vari√°veis

data_filtro = data.loc[:, ["cntry", "nwspol", "netustm", "ppltrst", "polintr",
                           "psppsgva", "trstprl", "trstlgl", "trstplt",
                           "trstep", "vote", "stfgov", "stfdem", "happy",
                           "gndr", "agea"]]
data_filtro
```

```python
# Limpando os dados inv√°lidos das vari√°veis

data_filtro = data_filtro.loc[(data_filtro["nwspol"] != 7777) &
                              (data_filtro["nwspol"] != 8888) &
                              (data_filtro["nwspol"] != 9999)]
data_filtro = data_filtro.loc[(data_filtro["netustm"] != 6666) &
                              (data_filtro["netustm"] != 7777) &
                              (data_filtro["netustm"] != 8888) &
                              (data_filtro["netustm"] != 9999)]
data_filtro = data_filtro.loc[(data_filtro["ppltrst"] != 77) &
                              (data_filtro["ppltrst"] != 88) &
                              (data_filtro["ppltrst"] != 99)]
data_filtro = data_filtro.loc[(data_filtro["polintr"] != 7) &
                              (data_filtro["polintr"] != 8) &
                              (data_filtro["polintr"] != 9)]
data_filtro = data_filtro.loc[(data_filtro["psppsgva"] != 7) &
                              (data_filtro["psppsgva"] != 8) &
                              (data_filtro["psppsgva"] != 9)]
data_filtro = data_filtro.loc[(data_filtro["trstprl"] != 77) &
                              (data_filtro["trstprl"] != 88) &
                              (data_filtro["trstprl"] != 99)]
data_filtro = data_filtro.loc[(data_filtro["trstlgl"] != 77) &
                              (data_filtro["trstlgl"] != 88) &
                              (data_filtro["trstlgl"] != 99)]
data_filtro = data_filtro.loc[(data_filtro["trstplt"] != 77) &
                              (data_filtro["trstplt"] != 88) &
                              (data_filtro["trstplt"] != 99)]
data_filtro = data_filtro.loc[(data_filtro["trstep"] != 77) &
                              (data_filtro["trstep"] != 88) &
                              (data_filtro["trstep"] != 99)]
data_filtro = data_filtro.loc[(data_filtro["vote"] != 7) &
                              (data_filtro["vote"] != 8) &
                              (data_filtro["vote"] != 9)]
data_filtro = data_filtro.loc[(data_filtro["stfgov"] != 77) &
                              (data_filtro["stfgov"] != 88) &
                              (data_filtro["stfgov"] != 99)]
data_filtro = data_filtro.loc[(data_filtro["stfdem"] != 77) &
                              (data_filtro["stfdem"] != 88) &
                              (data_filtro["stfdem"] != 99)]
data_filtro = data_filtro.loc[(data_filtro["happy"] != 77) &
                              (data_filtro["happy"] != 88) &
                              (data_filtro["happy"] != 99)]
data_filtro
```

```python
# Importando os pacotes.

from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from scipy.stats import ks_2samp
from sklearn.metrics import roc_curve, auc
```

```python
# Escolhendo a vari√°vel target e preparando os nossos dados.

X = data_filtro.loc[ : , data_filtro.columns!='trstplt']
y = data_filtro.trstplt
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,
                                                    random_state = 24)
print(f"Train Data: {X_train.shape}, {y_train.shape}")
print(f"Train Data: {X_test.shape}, {y_test.shape}")

classifier = RandomForestClassifier(random_state=18)
classifier.fit(X_train, y_train)

predictions = classifier.predict(X_test)
```
